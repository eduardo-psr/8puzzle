```python
#Definimos la classe Puzzle y todos los métodos que contiene, recibe una matriz nxn como entrada 
class Puzzle:
    def __init__(self, inicial) -> None:
        self.lista = []
        for i in inicial:          #Crea una lista a partir de la matriz de entrada
            for j in i:
                self.lista.append(j) 
        self.n = len(inicial)       #Determina n, la dimensión del tablero
        #Recupera la pieza del tablero a partir de las coordenadas
        self.pieza = lambda x: self.board[x[0]][x[1]]
        # Funciones lambda que devuelven las piezas que rodean a cierta pieza
        self.piece_up = lambda x: self.pieza([index_of(x, self.board)[0] - 1, index_of(x, self.board)[1]]) 
        self.piece_down = lambda x: self.pieza([index_of(x, self.board)[0] + 1, index_of(x, self.board)[1]]) 
        self.piece_left = lambda x: self.pieza([index_of(x, self.board)[0], index_of(x, self.board)[1] - 1])
        self.piece_right = lambda x: self.pieza([index_of(x, self.board)[0], index_of(x, self.board)[1] + 1])
        #Funciones lambda con operadores ternarios que determinan si existen bordes alrededor de una pieza
        self.down_border = lambda x: True if (index_of(x, self.board)[0] == self.n - 1) else False
        self.up_border = lambda x: True if index_of(x, self.board)[0] == 0 else False
        self.left_border = lambda x: True if index_of(x, self.board)[1] % self.n == 0 else False
        self.right_border = lambda x: True if (index_of(x, self.board)[1] + 1) % self.n == 0 else False
        #Variable que determina si un movimiento es possible o no lo es
        self.possible = False
        self.action = "" 
    def llenar(self):    
        """Crea un tablero de nxn a partir de la lista"""
        self.board = [self.lista[i * self.n: i * self.n + self.n] for i in range(self.n)]   
    def swap(self, a, b):
        """Función que intercambia dos elementos de la lista y actualiza el tablero"""
        self.lista[a], self.lista[b] = self.lista[b], self.lista[a]
        self.llenar()
        self.possible = True
    def mover(self, x):
        """Define los movimientos posibles de una pieza específica del tablero de acuerdo a su posición"""
        #esquina superior izquierda
        if self.up_border(x) and self.left_border(x):
            if self.piece_down(x) == "e":
                self.action = "Abajo"
                self.swap(self.lista.index(x), self.lista.index("e"))
            elif self.piece_right(x) == "e":
                self.action = "Derecha"
                self.swap(self.lista.index(x), self.lista.index("e"))
            else:
                self.action = "No se puede mover"
                self.possible = False
        #esquina superior derecha
        elif self.up_border(x) and self.right_border(x):
            if self.piece_down(x) == "e":
                self.action = "Abajo"
                self.swap(self.lista.index(x), self.lista.index("e"))
            elif self.piece_left(x) == "e":
                self.action = "Izquierda"
                self.swap(self.lista.index(x), self.lista.index("e"))
            else:
                self.action = "No se puede mover"
                self.possible = False
        #arista superior sin esquinas
        elif self.up_border(x):
            if self.piece_down(x) == "e":
                self.action = "Abajo"
                self.swap(self.lista.index(x), self.lista.index("e"))
            elif self.piece_right(x) == "e":
                self.action = "Derecha"
                self.swap(self.lista.index(x), self.lista.index("e"))
            elif self.piece_left(x) == "e":
                self.action = "Izquierda"
                self.swap(self.lista.index(x), self.lista.index("e"))
            else:
                self.action = "No se puede mover"
                self.possible = False
        #esquina inferior derecha
        elif self.right_border(x) and self.down_border(x):
            if self.piece_up(x) == "e":
                self.action = "Arriba"
                self.swap(self.lista.index(x), self.lista.index("e"))
            elif self.piece_left(x) == "e":
                self.action = "Izquierda"
                self.swap(self.lista.index(x), self.lista.index("e"))
            else:
                self.action = "No se puede mover"
                self.possible = False
        #esquina inferior izquierda
        elif self.left_border(x) and self.down_border(x):
            if self.piece_up(x) == "e":
                self.action = "Arriba"
                self.swap(self.lista.index(x), self.lista.index("e"))
            elif self.piece_right(x) == "e":
                self.action = "Derecha"
                self.swap(self.lista.index(x), self.lista.index("e"))
            else:
                self.action = "No se puede mover"
                self.possible = False
        #arista inferior sin esquinas 
        elif self.down_border(x):
            if self.piece_up(x) == "e":
                self.action = "Arriba"
                self.swap(self.lista.index(x), self.lista.index("e"))
            elif self.piece_left(x) == "e":
                self.action = "Izquierda"
                self.swap(self.lista.index(x), self.lista.index("e"))
            elif self.piece_right(x) == "e":
                self.action = "Derecha"
                self.swap(self.lista.index(x), self.lista.index("e"))
            else:
                self.action = "No se puede mover"
                self.possible = False
        #arista derecha sin esquinas
        elif self.right_border(x):
            if self.piece_up(x) == "e":
                self.action = "Arriba"
                self.swap(self.lista.index(x), self.lista.index("e"))
            elif self.piece_down(x) == "e":
                self.action = "Abajo"
                self.swap(self.lista.index(x), self.lista.index("e"))
            elif self.piece_left(x) == "e":
                self.action = "Izquierda"
                self.swap(self.lista.index(x), self.lista.index("e"))
            else:
                self.action = "No se puede mover"
                self.possible = False       
        #arista izquierda sin esquinas
        elif self.left_border(x):
            if self.piece_up(x) == "e":
                self.action = "Arriba"
                self.swap(self.lista.index(x), self.lista.index("e"))
            elif self.piece_down(x) == "e":
                self.action = "Abajo"
                self.swap(self.lista.index(x), self.lista.index("e"))
            elif self.piece_right(x) == "e":
                self.action = "Derecha"
                self.swap(self.lista.index(x), self.lista.index("e"))
            else:
                self.action = "No se puede mover"    
                self.possible = False    
        #pieza central
        else:
            if self.piece_up(x) == "e":
                self.action = "Arriba"
                self.swap(self.lista.index(x), self.lista.index("e"))
            elif self.piece_down(x) == "e":
                self.action = "Abajo"
                self.swap(self.lista.index(x), self.lista.index("e"))
            elif self.piece_left(x) == "e":
                self.action = "Izquierda"
                self.swap(self.lista.index(x), self.lista.index("e"))
            elif self.piece_right(x) == "e":
                self.action = "Derecha"
                self.swap(self.lista.index(x), self.lista.index("e"))
            else:
                self.action = "No se puede mover"
                self.possible = False

#Obtiene los indices de la pieza x en el tablero
def index_of(x, board):
    for i, board_i in enumerate(board):
        for j, value in enumerate(board_i):
            if value == x:
                return [i, j]

```

```python
# Implementación de Depth First Search
#Primero defino algunas funciones que usaré para la búsqueda
import os
def merge(x, y):
    """Función para combinar dos listas de tuplas sin estados duplicados"""
    sol = x
    aux = [i[1] for i in x]
    for i in y:
        if i[1] not in aux:
            sol.append(i)
    del aux
    return sol

def filter_list(listita, filtro):
    """Elimina los elementos de una lista que son elementos de la lista filtro"""
    aux = [i[1] for i in filtro]
    sol = [k for k in listita if k[1] not in aux]
    del aux
    return sol

def expandir(estado):
    """Función que devuelve una lista de tuplas (movimientos, estados) a partir de un tablero"""
    lista = []
    tablero = Puzzle(estado)
    tablero.llenar()
    for i in tablero.lista:
        tablero.mover(i)
        if tablero.possible:
            lista.append((tablero.action, tablero.board))
        tablero.mover(i)
    return lista

def dfs(inicial, objetivo, maximo):
    """Función que implementa Depth First Search"""
    goal = objetivo
    estado_inicial = ("inicio", inicial)
    current_node = ("inicio", inicial)
    frontier = [] 
    frontier = merge(frontier, expandir(current_node[1]))
    aux_frontier = []
    visited = []
    visited.append(current_node)
    iter = 1
    iter_max = maximo
    file = "salida.txt"            
    if os.path.exists(file):
        os.remove(file)
    with open(file, 'a') as f:
        f.write("iteracion: " + str(iter) + "\n")
        f.write("nodo actual: " + str(current_node) + "\n")
        f.write("tamanio frontera: " + str(len(frontier)) + "\n")
        f.write("tamanio visitados: " + str(len(visited))+ "\n")
    iter += 1
    current_node = frontier.pop()
    with open(file, "a") as f:
        f.write("nodo alcanzado: " + str(current_node) + "\n" + "\n")
    aux_frontier = expandir(current_node[1])
    while iter < iter_max and current_node[1] != goal:
        visited = merge(visited, [current_node])
        aux_frontier = filter_list(aux_frontier, visited)
        frontier = merge(frontier, aux_frontier)
        frontier = filter_list(frontier, [estado_inicial])
        with open(file, 'a') as f:
            f.write("iteracion: " + str(iter) + "\n")
            f.write("nodo actual: " + str(current_node) + "\n")
            f.write("tamanio frontera: " + str(len(frontier)) + "\n")
            f.write("tamanio visitados: " + str(len(visited))+ "\n")
        iter += 1
        current_node = frontier.pop()
        aux_frontier = expandir(current_node[1])
        with open(file, "a") as f:
            f.write("nodo alcanzado: " + str(current_node) + "\n" + "\n")


#Este es el número máximo de iteraciones que se realizaran en el algoritmo de búsqueda.
maximo_iteraciones = 7001


# Aquí hay un par de estados (inicial y final) como ejemplos para usar en la función dps, quitar el comentario del
# tablero que se quiera usar o colocar el tablero que se desee en forma de una matriz nxn adecuada para usar como instancia de Puzzle

#tablero_inicial = [[2, "e"],[1, 3]]   
#tablero_objetivo = [[1, 2], [3, "e"]]

tablero_inicial = [[1, "e", 2], [6, 3, 4], [7, 5, 8]]
tablero_objetivo = [[1, 2, 3], [4, 5, 6], [7, 8, "e"]]


```
